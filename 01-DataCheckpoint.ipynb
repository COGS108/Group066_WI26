## Research Question

Using statistical inference, what are the top predictive features among quantitative game statistics—such as total ratings, install milestones, and growth rates—that determine a game's specific rank within the Google Play Store’s genre-specific top 100 charts? Furthermore, does the relative importance of these features vary significantly across different game genres? 

**Target variable**   
- **rank** (ordinal/numerical: the game’s rank (1–100) in the genre’s top 100 list) 

**Predictor variables** 
- **total ratings** (numerical: game’s total number of ratings) 
- **installs** (ordinal/numerical: game’s approximate install milestone—e.g., 100.0 M installs vs. 500.0 M installs) 
- **average rating** (numerical: average rating out of 5)  
- **growth (30 days)** (numerical: percent growth in 30 days)  
- **growth (60 days)** (numerical: percent growth in 60 days) 
- **price** (numerical: price in dollars)   
- **5 star ratings** (numerical: number of 5 star ratings) 

**Grouping variable**

## Background and Prior Work

The Google Play Store is the top distributor of downloadable content for Android devices. Run by Google as Android's official app store, it offers apps, games, books, movies, TV, and other media<sup><a href="#ref1">1</a></sup>. As of 2024, the Google Play Store offered approximately 264,000 mobile games ranging across genres like Action, Puzzle, and RPG<sup><a href="#ref2">2</a></sup>. Within this ecosystem, top chart rankings play an important role in influencing which games users actually see and download. Higher ranked games receive more visibility in lists, search results, and curated sections, which in turn increases the likelihood that users will click through and install them<sup><a href="#ref3">3</a></sup>. However, since Google does not disclose its full algorithm behind rankings, it remains unclear which measurable performance metrics, such as install volume, rating averages, or review counts, most strongly predict a game's chart position<sup><a href="#ref3">3</a></sup>.

From our own experience scrolling through the Play Store, we noticed that we almost always check the star rating and number of reviews before downloading a game. If an app has a high average rating and thousands of reviews, it just feels more trustworthy than one with barely any feedback. So, we hypothesize that review count and average star rating will be the features most strongly correlated with a game's ranking. With hundreds of thousands of games competing for attention on the Play Store, pinpointing the very factors that increase the attraction of games will help game developers to focus their resources on improving in those specific areas instead of guessing what might work. This gap motivates our analysis: We seek to identify which quantitative signals in the Play Store data correlate most closely with ranking, thereby shedding light on the factors that likely drive discoverability in this marketplace. The "Top Games on Google Play Store" dataset provides rankings for the top 100 games across multiple genres, allowing us to analyze whether the relative importance of these predictive factors varies significantly between categories like Action, Puzzle, and RPG.

Prior studies indicate that publishers' prior release volume, intragenre ranking, consumer ratings, and review volume jointly drive high levels of game downloads, with the influence of any single information cue varying depending on the presence of other coexisting cues<sup><a href="#ref4">4</a></sup>. Complementary research using TAM and IDT based feature selection methods further shows that mobile game downloads are shaped by both usability perceptions, such as perceived ease of use and usefulness, and diffusion-related factors, including visibility, trialability, and social communication channels<sup><a href="#ref5">5</a></sup>. Together, these findings suggest that download performance emerges from interacting quantitative and contextual signals, a premise that directly informs our investigation of how such signals translate into genre specific ranking outcomes on the Google Play Store and similar app distribution platforms.

### Descriptions of any dataset shortcomings relevant to the project

This dataset has several limitations that may affect statistical analysis. First, installs are recorded as milestone categories (e.g., “1M+” or “10M+”) rather than exact counts, which reduces precision when converting them into numeric values for modeling. Second, the dataset is cross-sectional and represents only a single snapshot in time, so it cannot capture dynamic changes in rank, installs, or growth, nor establish causal relationships.  

There is also potential multicollinearity between installs and total ratings, since games with more downloads tend to accumulate more ratings. This may complicate regression interpretation and affect the relative importance of predictors. Finally, the dataset does not include other important ranking factors, such as revenue or user retention, which could result in omitted variable bias. Despite these limitations, the dataset is suitable for exploring predictive features of rank within the genre-specific Top 100 charts, which aligns with the research question.

### Top Games on Google Play Store 
This dataset contains the top 100 ranked Android games from each of 17 Google Play Store game categories (Action, Puzzle, Casino, etc.), totaling 1,730 rows before cleaning.

**Total ratings** is an integer count of how many users have left a star rating for the game. In this dataset it ranges from ~33,000 to ~86 million (Garena Free Fire). The median is about 430,000. To put that in context, a game with 100K ratings is reasonably well-known; a game with 1M+ ratings is a mainstream hit; and anything above 10M is a global phenomenon (e.g. PUBG, Free Fire, Subway Surfers). This metric is the best proxy we have for a game's overall popularity and user engagement, since actual download counts are bucketed.

**Installs** are reported as coarse string buckets — "100.0 k", "1.0 M", "10.0 M", "100.0 M", etc. — with only 9 distinct levels across the whole dataset. Nearly half of all games (47%) sit in the 10.0 M bucket, so this variable has extremely low resolution (i.e. a game listed as "10.0 M" installs could have anywhere from 10 million to 49.9 million actual installs).|

**Average rating** is reported as an integer — the only values present are 2, 3, and 4. This is the Play Store's displayed rating rounded to the nearest whole number, which destroys almost all signals (91% of games show a 4, 9% show a 3, and exactly one game shows a 2). This column is essentially useless for any comparative analysis.

**Growth (30 days) and growth (60 days)** represent the percentage increase in total ratings over the last 30 and 60 days, respectively. Typical values are small: the median 30-day growth is just 0.5%, reflecting the steady trickle of ratings that established games receive. Values of 5–10% suggest a game gaining meaningful momentum (perhaps from a marketing push). However, a handful of entries show absurd values—one game reports 227,106% 30-day growth—which almost certainly means it went from near-zero ratings to some modest number, making the percentage astronomically large despite a small absolute change.

**Star rating breakdowns (5-star through 1-star counts)** give the raw count of ratings at each level. These are highly correlated with total ratings (r > 0.93 for all five), since bigger games simply get more of everything. What's more analytically interesting is the ratio — for instance, a game where 1-star ratings make up 15%+ of its total (like Garena Free Fire at ~14.5%) signals polarizing quality or aggressive monetization that frustrates a vocal minority, even if the weighted average looks fine.

**The price** is in US dollars. The overwhelming majority of games (99.6%) are free; only 7 paid games appear in the dataset, ranging from \\$0.99 to \$7.49. This reflects the reality of the mobile gaming market: the top charts are dominated by free-to-play titles that monetize through ads and in-app purchases rather than upfront cost.

A key consideration for this dataset is that it includes only games ranked within the Google Play Store’s genre-specific Top 100 charts, which introduces selection bias by excluding lower-ranked or newly released games. As a result, the analysis focuses on explaining relative rank among already successful titles rather than general game success. Because the dataset contains only high-performing games, several quantitative features relevant to the research question, such as average rating and install milestones, exhibit limited variation, which may reduce their predictive power in statistical inference. Additionally, while rank values are expected to be unique within each genre, the dataset contains repeated rank numbers (e.g., multiple entries with the same rank such as 50), creating ambiguity in how rank is assigned and interpreted. Without additional information, it is unclear whether these repeated ranks correspond to ties, separate genre-specific rankings, or another ranking mechanism, which may affect the precision of rank-based modeling. Finally, the Play Store ranking system may incorporate factors not captured in the dataset, such as proprietary algorithms or platform-driven promotion, meaning rank may not be determined solely by observable metrics like total ratings, installs, or growth rates. Despite these limitations, the dataset remains appropriate for analyzing the relative importance of quantitative features in predicting rank within genre-specific Top 100 charts, which is the focus of this study.

##### We have unstandardized install counts, which we will convert from e.g. 10.0 M -> 10000000

```python
df_clean['installs']
```

```python
def convert_installs_to_value(value):
    if pd.isna(value):
        return value
    
    value = str(value).strip()
    
    if value.endswith('M') or value.endswith('m'):
        return float(value[:-1])*1000000
    elif value.endswith('k') or value.endswith('K'):
        return float(value[:-1])*1000
    else:
        # If no suffix, assume it's already a number
        return float(value)
```

```python
# Apply to your column
df_clean['installs'] = df_clean['installs'].apply(convert_installs_to_value)
df_clean
```
